```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sb
import numpy as np
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('salary.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(20,8))
sb.countplot(df['YearsExperience'])      
```

```python
# data with x axis expierence
x=df.iloc[:,0:1].values
x
```

```python
# salary putting in y axis
y=df.iloc[:,-1].values
y
```

```python
# training and testing  set
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#calling linear regression
from sklearn.linear_model import LinearRegression
regr=LinearRegression()
```

```python
#now applying 
fito=regr.fit(train_x,train_y)
#             train_x.reshape(-1,1)
```

```python
#
predict=fito.predict(test_x)
```

```python
plt.figure(figsize=(20,8))
plt.scatter(train_x,train_y,label='exp vs sal',color='red')
plt.plot(train_x,fito.predict(train_x),color='green')
plt.xlabel('Experienxe')
plt.ylabel('salary')
plt.legend()
```

```python
plt.figure(figsize=(20,8))
plt.scatter(train_x,train_y,label='exp vs sal',color='red')
plt.plot(test_x,predict,color='green')
plt.xlabel('Experienxe')
plt.ylabel('salary')
plt.legend()
```
